(hint: pd.read_csv(....., converts={}))

```python
def f(x):
    x = float(x)
    if x < 25:
        return ("Low")
    elif x > 25:
        return ("High")

df = pd.read_csv('data/BostonHousing.csv', converters={'medv':f})
df.head()
```
